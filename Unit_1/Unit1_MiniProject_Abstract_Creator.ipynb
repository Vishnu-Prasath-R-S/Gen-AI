# 35) Scientific Paper Abstract Creator

## Objective
To automatically generate a concise research abstract by summarizing the
methodology and results sections of a scientific paper using transformer-based
summarization models.


```python
from transformers import pipeline #PES2UG23CS703
```

```python
#PES2UG23CS703
summarizer = pipeline(
    "summarization",
    model="facebook/bart-large-cnn"
)
```

```python
#PES2UG23CS703
# Source paper:
# "Personalized Federated Learning With Differential Privacy and Convergence Guarantee"
# Authors: Kang Wei, Jun Li, Chuan Ma, Ming Ding, Wen Chen, Jun Wu, Meixia Tao, H. Vincent Poor
# Published in: IEEE Transactions on Information Forensics and Security, 2023

paper_text = """
This paper proposes a Rényi Differential Privacy based Personalized Federated Learning (RDP-PFL) framework
that integrates meta-learning to generate personalized models for heterogeneous clients. The framework applies
differentially private stochastic gradient descent (DP-SGD) in both inner and outer update stages to prevent
information leakage during training. A privacy budget allocation scheme is designed using Rényi Differential
Privacy composition theory.

The authors derive theoretical convergence bounds for the proposed framework under both convex and non-convex
loss function assumptions. The analysis reveals that there exists an optimal trade-off between privacy budget,
number of communication rounds, and model size. Experiments conducted on MNIST and CIFAR-10 datasets using MLP,
CNN, and ResNet-18 models show that the proposed RDP-PFL approach achieves strong convergence performance while
maintaining formal privacy guarantees.
"""
```

```python
#PES2UG23CS703
abstract = summarizer(
    paper_text,
    max_length=150,
    min_length=80,
    do_sample=False
)
```

```python
#PES2UG23CS703
print("Generated Abstract:\n")
print(abstract[0]["summary_text"])
```

The generated abstract successfully summarizes the core methodology and results of the paper. Minor grammatical inaccuracies were observed, indicating that while transformer-based summarization is effective, human validation is still required for technical writing.
